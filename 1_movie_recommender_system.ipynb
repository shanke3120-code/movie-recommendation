### Merging the two Datasets on 'id'

```python
movies = movies.merge(credits, on='title')

movies.head(2)
```

### Choosing the cols

```python
### Choosing important cols only for recommending:
#### id
#### original_title
#### genres
#### keywords
#### overview
#### cast
####crew 
```

```python
movies = movies[["id_x", "title", "keywords", "genres", "overview", "cast", "crew"]]
movies.head(2)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

### Changing str of lists into lists

```python
import ast
ast.literal_eval(movies['genres'][0])
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])

    return L

movies["genres"] = movies['genres'].apply(convert)
movies["keywords"] = movies['keywords'].apply(convert)

movies.head()
```

```python
def take_cast(obj):
    L = []
    for i in ast.literal_eval(obj)[:3]:
        L.append(i['name'])
    return L

movies["cast"] = movies["cast"].apply(take_cast)
movies.head(2)
```

```python
def take_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            L.append(i["name"])
            break
    return L

movies["crew"] = movies["crew"].apply(take_director)
movies.head(2)
```

```python
movies["overview"][0]
```

### changes str into list

```python
movies["overview"] = movies["overview"].apply(lambda x: x.split())
```

```python
movies.head(3)
```

```python
movies["genres"] = movies["genres"].apply(lambda x: [i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x: [i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x: [i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x: [i.replace(" ","") for i in x])

movies.head(2)
```

### Creating new tag columns

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]

movies.head(3)
```

```python
movies["tags"]
```

```python
new_df = movies[["id_x", "title","tags"]]

new_df.rename(columns={"id_x" : "id"}, inplace=True)

new_df.head(3)
```

### craeting list into str in column Tag

```python
new_df["tags"] = new_df["tags"].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
new_df["tags"][0]
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x: x.lower())
```

```python
new_df.head(3)
```

## Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()

vectors
```

```python
import nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

new_df["tags"].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

## Main Function


```python
# First getting index of a movie
new_df[new_df["title"] == 'Avatar'].index[0]
```

```python
sorted(similarity[0], reverse=True)[0:10]
```

```python
# gives us list of tuples which gives distance of 0th movie from 0th, 1st, 2nd and like that......
list(enumerate(similarity[0]))[0:10]
```

```python
# Now sorting the distances of the movies based on index
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1])[1:6]
```

```python
# Getting the name of the movie based on index
new_df.iloc[1192].title
```

```python
def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Batman Begins")
```
